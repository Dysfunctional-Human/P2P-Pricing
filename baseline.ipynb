```python
# mvp_p2p_baseline.py
import pandapower as pp
import pandapower.networks as pn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ------------- parameters -------------
T = 24  # hours (extend to 8760 for year)
tariff_export = 3.0  # currency units / kWh paid by utility for exported energy (set to govt tariff)
retail_price = 8.0   # price paid by prosumer for imported energy
# sample prosumers: bus index -> basic params
prosumers = {
    3: {"pv_kw": 2.0, "load_kw": None},  # we'll populate time series
    7: {"pv_kw": 3.5, "load_kw": None},
}

# ------------- load network -------------
net = pn.case14()  # small IEEE case; choose case24 or others as needed

# ------------- sample time series (replace with real data) -------------
# generate toy daily profile
hours = np.arange(T)
pv_profile_norm = np.maximum(0, np.sin((hours - 6)/24 * 2*np.pi))  # simple bell
load_profile = 0.6 + 0.4 * (0.5 + 0.5*np.cos((hours-18)/24*2*np.pi))  # toy residential

# attach time series to prosumers
for bus, p in prosumers.items():
    p["pv_ts"] = p["pv_kw"] * pv_profile_norm  # kW per hour
    p["load_ts"] = load_profile * (1.0 + np.random.randn(T)*0.05)  # small noise

# bookkeeping
results = {bus: {"export_kwh": np.zeros(T), "import_kwh": np.zeros(T)} for bus in prosumers}

# ------------- simulation loop -------------
for t in range(T):
    # reset injections in net
    # We'll treat prosumers as loads (positive) and add gen (negative injection) or set ext_grid injections
    # For simplicity, modify net.load/power injection values for bus indexes.
    # NOTE: pandapower expects elements; simpler approach: use net.load and net.sgen entries added for prosumers.
    # Here we will remove previous sgen/load and re-add each timestep for clarity.
    # Clear existing prosumer loads/sgens if any
    net.load.drop(net.load[net.load['name'].fillna('').str.contains('prosumer_')].index, inplace=True)
    net.sgen.drop(net.sgen[net.sgen['name'].fillna('').str.contains('prosumer_')].index, inplace=True)
    
    for bus, p in prosumers.items():
        pv = p["pv_ts"][t]
        load = p["load_ts"][t]
        # dispatch: pv serves load first
        self_consumed = min(pv, load)
        export = max(0.0, pv - load)
        import_ = max(0.0, load - pv)
        # record
        results[bus]["export_kwh"][t] = export
        results[bus]["import_kwh"][t] = import_
        # add to pandapower model using concat
        load_row = pd.DataFrame({
            "bus": [int(bus)], "p_mw": [load/1000.0], "q_mvar": [0.0], "name": [f"prosumer_load_{bus}"], "in_service": [True]
        })
        net.load = pd.concat([net.load, load_row], ignore_index=True)
        net.load["bus"] = net.load["bus"].astype("int64")
        net.load["in_service"] = net.load["in_service"].astype("bool")
        
        sgen_row = pd.DataFrame({
            "bus": [int(bus)], "p_mw": [pv/1000.0], "q_mvar": [0.0], "name": [f"prosumer_pv_{bus}"], "in_service": [True]
        })
        net.sgen = pd.concat([net.sgen, sgen_row], ignore_index=True)
        net.sgen["bus"] = net.sgen["bus"].astype("int64")
        net.sgen["in_service"] = net.sgen["in_service"].astype("bool")
    # run powerflow
    try:
        pp.runpp(net)
        # inspect voltages/line loading if needed
    except pp.powerflow.LoadflowNotConverged:
        print(f"Power flow failed at timestep {t}.")
        # handle (flag) - for MVP we just continue

# ------------- postprocess -------------
summary = []
for bus, p in prosumers.items():
    total_export = results[bus]["export_kwh"].sum()
    total_import = results[bus]["import_kwh"].sum()
    revenue = total_export * tariff_export
    cost_import = total_import * retail_price
    net_bill = cost_import - revenue
    summary.append({
        "bus": bus,
        "total_export_kwh": total_export,
        "total_import_kwh": total_import,
        "revenue": revenue,
        "cost_import": cost_import,
        "net_bill": net_bill
    })
df_summary = pd.DataFrame(summary)
print(df_summary)

# plot one prosumer
bus0 = list(prosumers.keys())[0]
plt.figure()
plt.plot(hours, prosumers[bus0]["pv_ts"], label="pv_kW")
plt.plot(hours, prosumers[bus0]["load_ts"], label="load_kW")
plt.plot(hours, results[bus0]["export_kwh"], label="export_kW")
plt.legend()
plt.xlabel("hour")
plt.ylabel("kW")
plt.title(f"Prosumer at bus {bus0}")
plt.show()
```

```python
import pandapower as pp
import pandapower.networks as pn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------------- PARAMETERS ----------------
T = 24      # Number of time steps for simulation (in hours). Controls the length of time series arrays, and the number of loop iterations in the simulation loop.
tariff_export = 3.0       # rupees / kwh   (price set by government for buyng energy back from the prosumer)
retail_price = 8.0        # currency / kWh  (regular electricity buying price)

# choose test case
network = pn.case14()         # Loads the IEEE 14 bus system
try:
    pp.runpp(net, init="flat")      # test run the initial 14 bus system before adding prosumers to make sure it converges by default. "flat" sets initial voltages to 1 pu and 0 angle.
except Exception as e:
    print("Base case power flow failed on original case14:", e)

# sample prosumers at buses that exist in case14 (0..13)
# Current setup uses only 2 prosumers and buses 3 and 7 along with their installed PV generation capacity
prosumers = {
    3: {"pv_kw": 2.0},
    7: {"pv_kw": 3.5},
}

# ---------------- TIME SERIES (toy -> replace with real) ----------------
hours = np.arange(T)    # Creates a numpy array of time steps (0-23)
pv_profile_norm = np.maximum(0, np.sin((hours - 6)/24 * 2*np.pi))
load_profile = 0.6 + 0.4 * (0.5 + 0.5*np.cos((hours-18)/24*2*np.pi))

for bus, p in prosumers.items():
    p["pv_ts_kw"] = p["pv_kw"] * pv_profile_norm          # kW
    p["load_ts_kw"] = load_profile * (1.0 + np.random.randn(T)*0.05) * 1.5  # scale for realism

# ---------------- CREATE elements ONCE and store indices ----------------
# We'll add a prosumer load and a sgen for each prosumer and update p_mw every step
for bus, p in prosumers.items():
    # create a load (p_mw in MW)
    idx_load = pp.create_load(net, bus=bus, p_mw=0.0, q_mvar=0.0, name=f"prosumer_load_{bus}")
    idx_sgen = pp.create_sgen(net, bus=bus, p_mw=0.0, q_mvar=0.0, name=f"prosumer_pv_{bus}")
    p["load_idx"] = idx_load
    p["sgen_idx"] = idx_sgen

# bookkeeping
results = {bus: {"export_kwh": np.zeros(T), "import_kwh": np.zeros(T)} for bus in prosumers}
failed_steps = []

# ---------------- SIMULATION LOOP ----------------
for t in range(T):
    # update loads / sgens in-place (convert kW -> MW)
    for bus, p in prosumers.items():
        pv_kw = float(p["pv_ts_kw"][t])
        load_kw = float(p["load_ts_kw"][t])
        # dispatch locally (PV serves load first)
        self_consumed = min(pv_kw, load_kw)
        export_kw = max(0.0, pv_kw - load_kw)
        import_kw = max(0.0, load_kw - pv_kw)

        # record energy (kWh)
        results[bus]["export_kwh"][t] = export_kw
        results[bus]["import_kwh"][t] = import_kw

        # update pandapower element values (in MW)
        net.load.at[p["load_idx"], "p_mw"] = load_kw / 1000.0
        net.sgen.at[p["sgen_idx"], "p_mw"] = pv_kw / 1000.0

    # try running powerflow safely; catch exceptions for failed steps
    try:
        # `init="flat"` provides a stable initial guess in many cases
        # increase max_iteration or decrease tolerance if needed
        pp.runpp(net, init="flat", max_iteration=200, tol_kva=1e-4)
    except Exception as e:
        # record failed timestep
        failed_steps.append((t, str(e)))
        # optionally: try a recovery run with different init or solver options
        try:
            pp.runpp(net, init="results", max_iteration=200)
        except Exception as e2:
            # still failed -> continue but log
            failed_steps.append((t, "recovery_failed: " + str(e2)))
            # Option: zero-out the prosumer injections for next iterations to avoid repeated failure
            # for bus,p in prosumers.items():
            #     net.load.at[p["load_idx"], "p_mw"] = 0.0
            #     net.sgen.at[p["sgen_idx"], "p_mw"] = 0.0

# ---------------- POSTPROCESS ----------------
summary = []
for bus, p in prosumers.items():
    total_export = results[bus]["export_kwh"].sum()
    total_import = results[bus]["import_kwh"].sum()
    revenue = total_export * tariff_export
    cost_import = total_import * retail_price
    net_bill = cost_import - revenue
    summary.append({
        "bus": bus,
        "total_export_kwh": total_export,
        "total_import_kwh": total_import,
        "revenue": revenue,
        "cost_import": cost_import,
        "net_bill": net_bill
    })
df_summary = pd.DataFrame(summary)
print(df_summary)
print("Failed timesteps count:", len(failed_steps))
if failed_steps:
    print("Failed details (first 10):", failed_steps[:10])

# quick plot for first prosumer
bus0 = list(prosumers.keys())[0]
plt.figure(figsize=(8,4))
plt.plot(hours, prosumers[bus0]["pv_ts_kw"], label="pv_kW")
plt.plot(hours, prosumers[bus0]["load_ts_kw"], label="load_kW")
plt.plot(hours, results[bus0]["export_kwh"], label="export_kW")
plt.legend()
plt.xlabel("hour")
plt.ylabel("kW")
plt.title(f"Prosumer at bus {bus0}")
plt.tight_layout()
plt.show()
```

```python
def print_bus_snapshot(t, net, prosumers, results=None, show_residuals=True):
    """
    Print a per-bus snapshot at time index t.

    Parameters
    ----------
    t : int
        Timestep index (0..T-1).
    net : pandapowerNet
        The pandapower network object (must be the same `net` you update).
    prosumers : dict
        The prosumers dict used in your code (keys = bus indices).
        Each prosumer entry is expected to contain:
            - 'pv_ts_kw' (array-like, kW)
            - 'load_ts_kw' (array-like, kW)
            - 'load_idx' and 'sgen_idx' (optional) for direct net mapping
    results : dict, optional
        The results dict you build (like in your script). If provided and contains
        results[bus]['export_kwh'][t], that export value will be shown.
    show_residuals : bool
        If True, compute and show net_injection = generation - load (kW). Positive => export.

    Behavior / notes
    ----------------
    - Call this after you update net.load.p_mw and net.sgen.p_mw for timestep t (so values reflect current step).
    - It will inspect both pandapower net tables (net.load / net.sgen) and your `prosumers` time series; it prefers
      the time-series from `prosumers` for scheduled values when available (so you see exactly the pv/load_ts values).
    - For buses that are not in `prosumers`, the function reports aggregated loads/generation from net.load/net.sgen rows.
    - If `results` is supplied and contains per-bus export/import entries, those are printed too.
    """

    # Defensive checks
    if t is None:
        print("Error: timestep t must be provided (int)")
        return
    if not hasattr(net, "bus"):
        print("Error: the 'net' argument does not look like a pandapower network.")
        return

    rows = []
    for bus in net.bus.index:
        # scheduled (time-series) values if this bus is a prosumer in your dict
        scheduled_load_kw = None
        scheduled_pv_kw = None
        if bus in prosumers:
            p = prosumers[bus]
            try:
                scheduled_load_kw = float(p.get("load_ts_kw", np.zeros(1))[t])
            except Exception:
                scheduled_load_kw = None
            try:
                scheduled_pv_kw = float(p.get("pv_ts_kw", np.zeros(1))[t])
            except Exception:
                scheduled_pv_kw = None

        # current values in pandapower tables (p_mw stored in MW)
        load_rows = net.load.loc[net.load["bus"] == bus, "p_mw"]
        gen_rows = net.sgen.loc[net.sgen["bus"] == bus, "p_mw"]

        load_kw_net = float(load_rows.sum() * 1000.0) if not load_rows.empty else 0.0
        gen_kw_net = float(gen_rows.sum() * 1000.0) if not gen_rows.empty else 0.0

        # Prefer showing scheduled timeseries values (they are the source of truth for dispatch)
        # but also show the net table value so the user can compare
        load_display = scheduled_load_kw if scheduled_load_kw is not None else load_kw_net
        pv_display = scheduled_pv_kw if scheduled_pv_kw is not None else gen_kw_net

        # net injection (generation - load) from the pandapower tables (reflects what PF will see)
        net_injection_kw = gen_kw_net - load_kw_net if show_residuals else None

        # exported/imported from accounting results if provided
        export_kwh = None
        import_kwh = None
        if results is not None and bus in results:
            try:
                export_kwh = float(results[bus]["export_kwh"][t])
            except Exception:
                export_kwh = None
            try:
                import_kwh = float(results[bus]["import_kwh"][t])
            except Exception:
                import_kwh = None

        # Compose row
        rows.append({
            "bus": bus,
            "scheduled_load_kW": round(scheduled_load_kw, 4) if scheduled_load_kw is not None else None,
            "scheduled_pv_kW": round(scheduled_pv_kw, 4) if scheduled_pv_kw is not None else None,
            "net_table_load_kW": round(load_kw_net, 4),
            "net_table_gen_kW": round(gen_kw_net, 4),
            "net_injection_kW": round(net_injection_kw, 4) if net_injection_kw is not None else None,
            "export_kWh_recorded": round(export_kwh, 4) if export_kwh is not None else None,
            "import_kWh_recorded": round(import_kwh, 4) if import_kwh is not None else None,
        })

    # build a pandas DataFrame to display nicely
    df = pd.DataFrame(rows)
    # Order columns for clarity
    cols = ["bus", "scheduled_load_kW", "scheduled_pv_kW", "net_table_load_kW",
            "net_table_gen_kW", "net_injection_kW", "export_kWh_recorded", "import_kWh_recorded"]
    df = df[cols]

    # Print header and the table
    print(f"\n--- Bus snapshot at timestep t={t} ---")
    print(df.to_string(index=False))
    print("--- end snapshot ---\n")

    # If you want to return the DataFrame instead of (or as well as) printing, return it:
    return df
```
